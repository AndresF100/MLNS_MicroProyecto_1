```python
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
```

# Data

```python
def load_images(root_path, file_names, flatten=True, normalize=True):
    """
    Retorna una lista con arreglos de números. Cada arreglo representa una imagen RGB.
    
    Parametros:
    root_path : str
        Directorio raíz de las imágenes.
    file_names : list
        Lista con los nombres de los archivos de imágenes.
    flatten : bool, opcional
        Si es True, aplana las imágenes a una lista de píxeles de tres dimensiones.
    normalize : bool, opcional
        Si es True, normaliza los valores de píxeles al rango [0, 1].
    """
    image_list = list()
    for file in file_names:
        img = cv2.imread(os.path.join(root_path, file))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        if flatten:
            img = img.reshape((-1, 3))
        if normalize:
            img = img / 255.0
        image_list.append(img)
    return image_list


image_path = './Data/'
image_folders = os.listdir(image_path)

image_folders = [os.path.join(image_path, folder) for folder in image_folders]    

selected_folder = 0

image_list = os.listdir(image_folders[selected_folder])
image_list = load_images(image_folders[selected_folder], image_list,False,False)
```

```python
plt.figure(figsize=(10, 5))

plt.title('Imagen Original')
plt.imshow(image_list[1])
plt.axis('off')

plt.tight_layout()
plt.show()
```
